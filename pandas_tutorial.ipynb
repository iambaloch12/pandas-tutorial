**Conditional Selections(Boolean Arrays)**

```python
g7_pop < 70 #population less than 70 millions are true
```

```python
g7_pop[g7_pop > 70] # gives only which satisfy the condition
```

```python
g7_pop[g7_pop < 40]
```

```python
g7_pop.mean() #gives mean value of all population
```
